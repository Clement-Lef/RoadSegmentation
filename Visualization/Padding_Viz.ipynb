```python
import numpy as np

def mirror_padding(image, heigth, width):
        """ Pads the image using symmetric boundary conditions
            heigth : number of elements to pad on the right and bottom edges
            width  : number of elements to pad on the top and left edges
            Returns: Padded image
        """

        newShape = list(image.shape)
        newShape[0] += 2*heigth
        newShape[1] += 2*width

        newImage = np.empty(newShape)

        if len(image.shape)<3:
            newImage = np.pad(image[:,:], (heigth, width), 'symmetric')
        else:
            newImage = np.asarray([np.pad(image[:,:,i], (heigth, width), 'symmetric')\
                                  for i in range(image.shape[2])])
            newImage = np.moveaxis(newImage, 0, -1)

        return newImage
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
window_size = 64
patch_size = 16
pad_size = int((window_size-patch_size)/2)

corner_patch = (100,100)
corner_window = (corner_patch[0] - pad_size,corner_patch[1]-pad_size)

fig,ax = plt.subplots(1,2, figsize = (12,10))
im = plt.imread('datas/satImage_001_tr.png')
im_contour = plt.imread('datas/satImage_001_contour.png')
im_pad = mirror_padding(im,pad_size,pad_size)


implot1 = ax[0].imshow(im_contour)

implot2 = ax[1].imshow(im_pad)
rect = patches.Rectangle(corner_patch,patch_size,patch_size,fill = True, alpha = 0.9, color = 'red')
rect2 = patches.Rectangle(corner_window,window_size,window_size,color = 'blue', alpha = 0.5)
ax[1].add_patch(rect)
ax[1].add_patch(rect2)

ax[0].axis('off')
ax[1].axis('off')

plt.tight_layout()

plt.show()
fig.savefig('Example_Window.png')
```
